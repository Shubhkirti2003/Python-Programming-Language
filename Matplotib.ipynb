```python
names=["Aron",'Betty',"Cathy"]
marks=[87,58,98]

plt.figure(figsize=(5,10))

plt.subplot(131)
plt.bar(names, marks)
plt.subplot(132)
plt.scatter(names,marks)
plt.subplot(133)
plt.plot(names,marks)
plt.title("WWW")
plt.suptitle('Categorical plotting')
plt.show()
```
